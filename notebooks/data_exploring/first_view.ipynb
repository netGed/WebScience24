2. Clean:
2.1 Datenvorverarbeitung:
Die Vorverarbeitung ist entscheidend, um die Textdaten vor der Analyse zu bereinigen und zu normalisieren. Dieser Schritt umfasst mehrere Unteraufgaben:

- Tokenisierung: Aufteilen des Textes in einzelne Wörter oder Token. Beispielsweise wird aus „Hallo Welt!“ [„Hallo“, „Welt“, „!“].
- irrelevant - Kleinschreibung: Konvertieren aller Zeichen in Kleinbuchstaben, um Einheitlichkeit zu gewährleisten (z. B. werden „Hallo“ und „hallo“ gleich behandelt).
- Entfernen von Satzzeichen: Eliminieren von Satzzeichen, da diese oft nicht zur Bedeutung beitragen.
- Entfernen von Stoppwörtern: Entfernen gebräuchlicher Wörter wie „und“, „das“, „ist“, die keine bedeutende Bedeutung haben.
- Lemmatisierung: Konvertieren von Wörtern in ihre Grund- oder Stammform (z. B. wird „laufen“ zu „laufen“).
- Stemming: Ähnlich wie Lemmatisierung, aber normalerweise aggressiver. Dabei werden Wörter auf ihren Stamm reduziert (z. B. wird „laufen“ zu „laufen“, „Läufer“ zu „laufen“).

2.1. Merkmalsextraktion
Dieser Schritt wandelt den Text in numerische Merkmale um, die von maschinellen Lernmodellen verwendet werden können.

Bag of Words (BoW): Stellt Text als eine Sammlung von Wörtern unabhängig von Grammatik und Reihenfolge dar und gibt jedem Wort einen eindeutigen Index.
TF-IDF (Term Frequency-Inverse Document Frequency): Misst die Wichtigkeit eines Wortes in einem Dokument im Verhältnis zum gesamten Korpus.
Word Embeddings: Wandelt Wörter in dichte Vektoren um, die semantische Bedeutung erfassen. Beispiele sind Word2Vec, GloVe und FastText.

3. Use/Reuse
3.1 Modelltraining
Mit den im vorherigen Schritt extrahierten Merkmalen können Sie nun maschinelle Lernmodelle trainieren. Dies könnte verschiedene NLP-Aufgaben umfassen, wie z. B.:

Textklassifizierung: Zuweisen von Kategorien zu Text (z. B. Spam-Erkennung, Stimmungsanalyse).
Named Entity Recognition (NER): Identifizieren von Entitäten wie Namen, Daten und Orten im Text.
Wortarten-Tagging (POS): Identifizieren grammatikalischer Wortarten (Substantive, Verben, Adjektive usw.).

Abhängigkeitsanalyse: Analysieren der grammatikalischen Struktur eines Satzes.

Maschinelle Übersetzung: Übersetzen von Text von einer Sprache in eine andere.

Textzusammenfassung: Verdichten eines Textes in eine kürzere Version unter Beibehaltung der Hauptpunkte.

4.1 Modellbewertung
Bewerten Sie die Leistung des trainierten Modells anhand von Kennzahlen wie Genauigkeit, Präzision, Rückruf, F1-Score usw. Dies hilft zu verstehen, wie gut das Modell bei unbekannten Daten funktioniert.
