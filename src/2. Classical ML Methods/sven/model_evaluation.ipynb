```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tfidf = pd.read_csv('./eval_data/nb_grid_tfidf.csv')#, index_col=0)
tfidf_cnb = pd.read_csv('./eval_data/cnb_grid_tfidf.csv')#, index_col=0)
w2v = pd.read_csv('./eval_data/nb_grid_w2v.csv')#, index_col=0)
w2v_param = pd.read_csv('./eval_data/nb_grid_w2v_parameters.csv')#, index_col=0)
glv = pd.read_csv('./eval_data/nb_grid_glv.csv')#, index_col=0)
```

```python
evaluation = pd.concat([tfidf,
                        tfidf_cnb,
                        w2v,
                        w2v_param,
                        glv,
                        ])
```

```python
evaluation
```

```python
evaluation.sort_values(by=["test_recall"], ascending=False)
```

```python
evaluation.sort_values(by=["test_precision"], ascending=False)
```

```python
evaluation.sort_values(by=["test_f1"], ascending=False)
```

### Visualisierung F1-Score

```python
fig, ax = plt.subplots(figsize=(20, 7))
plt.xticks(rotation=30)
sns.barplot(data=evaluation, y=evaluation.test_f1, x=evaluation.model, label="Test F1-Score")
```

### Unterschiede/Schwankungen zw. Train <-> Test F1-Scores

```python
fig, ax = plt.subplots(figsize=(20, 7))
plt.xticks(rotation=30)

diff = evaluation['train_f1'] - evaluation['test_f1']
sns.scatterplot(data=evaluation, y=evaluation['train_f1'], x=evaluation.model, ax=ax, label="Train F1-Score")
sns.scatterplot(data=evaluation, y=evaluation['test_f1'], x=evaluation.model, ax=ax, label="Test F1-Score")
sns.lineplot(data=evaluation, y=diff, x=evaluation.model, label="Differenz Train-Test")

plt.show()
```

### Vergleich Accuracy - F1-Score

```python
fig, ax = plt.subplots(figsize=(20, 7))
plt.xticks(rotation=30)
sns.barplot(data=evaluation, y=evaluation.test_f1, x=evaluation.model, label="F1-Score")
sns.lineplot(data=evaluation, y=evaluation.test_accuracy, x=evaluation.model, label="Accuracy")
```

### Vergleich Precision - Recall

```python
fig, ax = plt.subplots(figsize=(20, 7))
plt.xticks(rotation=30)
a = sns.lineplot(data=evaluation, y=evaluation.test_precision, x=evaluation.model, label="Precision")
b = sns.lineplot(data=evaluation, y=evaluation.test_recall, x=evaluation.model, label="Recall")
```

#### Auswertung

- Bester Recall mit Glove, dafür aber sehr schlechte Precision
- W2V ähnlich gut im Recall, dabei etwas besser in Accuracy und F1-Score (aber immer noch schlecht)
- Beste Precision mit TF-IDF (Multi), aber schlechter Recall
- Bester F1-Score mit TF-IDF (Complement), Precision, Recall und Accuracy zwischen 0,4 und 0,6 --> ausgewogenestes Model
- Unterschied Performance Train/Test: TF-IDF auf Test-Daten deutlich schlechter, W2V und Glove ähnliches Verhalten Train/Test

--> Precision und Recall verhalten sich gegenläufig in den Modellen
