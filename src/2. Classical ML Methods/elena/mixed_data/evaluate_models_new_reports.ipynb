```python
import os
import pandas as pd
```

```python
#loading files and bringing them togehter in one df
file_paths = [
    # tfidf
    './tfidf/3_Lauf_neue_reports/linear/svm_tfidf_linear_results.csv',
    './tfidf/3_Lauf_neue_reports/poly/svm_tfidf_poly_results.csv',
    './tfidf/3_Lauf_neue_reports/rbf/svm_tfidf_rbf_results.csv',
    './tfidf/3_Lauf_neue_reports/sigmoid/svm_tfidf_sigmoid_results.csv',
    # w2v
    './w2v/3_Lauf_neue_reports/linear/svm_w2v_linear_results.csv',
    './w2v/3_Lauf_neue_reports/poly/svm_w2v_poly_results.csv',
    './w2v/3_Lauf_neue_reports/rbf/svm_w2v_rbf_results.csv',
    './w2v/3_Lauf_neue_reports/sigmoid/svm_w2v_sigmoid_results.csv',
    # glove
    './glove/B_Lauf_neue_reports/linear/svm_glove_linear_results.csv',
    './glove/B_Lauf_neue_reports/poly/svm_glove_poly_results.csv',
    './glove/B_Lauf_neue_reports/rbf/svm_glove_rbf_results.csv',
    './glove/B_Lauf_neue_reports/sigmoid/svm_glove_sigmoid_results.csv'
]

dataframes = []

for file_path in file_paths:
    file_name = os.path.basename(file_path) 
    df = pd.read_csv(file_path)
    df['source_file'] = file_name  
    dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
# save resulting df
combined_df.to_csv('combined_output.csv', index=False)
```
