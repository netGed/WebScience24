```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tfidf_lin = pd.read_csv('./eval_data/svm_grid_tfidf_linear.csv', index_col=0)
tfidf_sig = pd.read_csv('./eval_data/svm_grid_tfidf_sigmoid.csv', index_col=0)
tfidf_rbf = pd.read_csv('./eval_data/svm_grid_tfidf_rbf.csv', index_col=0)
tfidf_poly = pd.read_csv('./eval_data/svm_grid_tfidf_poly.csv', index_col=0)
w2v_lin = pd.read_csv('./eval_data/svm_grid_w2v_linear.csv', index_col=0)
w2v_sig = pd.read_csv('./eval_data/svm_grid_w2v_sigmoid.csv', index_col=0)
w2v_rbf = pd.read_csv('./eval_data/svm_grid_w2v_rbf.csv', index_col=0)
w2v_poly = pd.read_csv('./eval_data/svm_grid_w2v_poly.csv', index_col=0)
```

```python
evaluation = pd.concat([tfidf_lin,
                        tfidf_sig,
                        tfidf_rbf,
                        tfidf_poly,
                        w2v_lin, 
                        w2v_sig,
                        w2v_rbf, 
                        w2v_poly
                        ])

evaluation.insert(0, 'ID', range(1, len(evaluation) + 1))
```

#### Visualisierung F1-Score

```python
vis_data = evaluation.sort_values(by=["test_f1"], ascending=False).head(50)

fig, ax = plt.subplots(figsize=(20, 7))
plt.xticks(rotation=30)
sns.barplot(data=vis_data, y=vis_data.test_f1, x=vis_data.ID, label="Test F1-Score")
```

#### Unterschiede/Schwankungen zw. Train <-> Test F1-Scores

```python
fig, ax = plt.subplots(figsize=(20, 7))
plt.xticks(rotation=30)

evaluation['diff'] = evaluation['train_f1'] - evaluation['test_f1']
evaluation['diff'] = evaluation['diff'].replace([float('inf'), -float('inf')], pd.NA)
sns.scatterplot(data=evaluation, y=evaluation['train_f1'], x=evaluation.ID, ax=ax, label="Train F1-Score")
sns.scatterplot(data=evaluation, y=evaluation['test_f1'], x=evaluation.ID, ax=ax, label="Test F1-Score")
sns.lineplot(data=evaluation, y='diff', x='ID', label="Differenz Train-Test")

plt.show()


# # Assuming 'evaluation' is your DataFrame
# evaluation['diff'] = evaluation['diff'].replace([float('inf'), -float('inf')], pd.NA)

# # Now, plot the data
# sns.lineplot(data=evaluation, y='diff', x='ID', label="Differenz Train-Test")



```
