```python
color_mapping = {
    'Negative': 'red',
    'Positive': 'green',
    'Neutral': 'orange',
    'Irrelevant': 'grey'
}

counts = df['HateCategorization'].value_counts()

colors = counts.index.map(color_mapping)

ax = counts.plot(kind='barh', color=colors)

for p in ax.patches:
    ax.annotate(str(p.get_width()), (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='left', va='center', xytext=(5, 0), textcoords='offset points')

ax.set_xlim(0, max(counts) * 1.1)

plt.tight_layout()

plt.show()
```
