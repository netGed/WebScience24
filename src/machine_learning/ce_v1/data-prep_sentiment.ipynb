```python
ax = df['label'].value_counts().plot(kind='barh')

for p in ax.patches:
    ax.annotate(str(p.get_width()), (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='left', va='center', xytext=(5, 0), textcoords='offset points')

ax.set_xlim(0, max(df['label'].value_counts()) * 1.1)

plt.tight_layout()
    
plt.show()
```
